```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('tweets.csv')
df.head(5)
```

```python
df = df.drop(columns=['longitude', 'latitude', 'country'])
df.insert(4, "Date", df["date_time"].str.split(" ").str[0])
df.insert(4, "Time", df["date_time"].str.split(" ").str[-1])
df = df.drop(columns=['date_time'])
df['language'] = df['language'].str.upper()

df.head(5)
```

```python
# Who has received a lot more interation through posts overall
dfAna = df.drop(columns=['id','content', 'Time', 'Date', 'language'])
dfAna = dfAna.groupby(['author']).sum().sort_values(by=['number_of_shares'], ascending=False)
dfAna
```

```python
savedData = df[df['author'] == 'katyperry']
savedData
```

```python
savedData.to_csv('katyperry_twitter_data.csv')
dfAna.to_csv('twitter_likes_data.csv')
```
